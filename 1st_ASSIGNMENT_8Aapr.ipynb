## **Q9: You are building a simple sales analysis system.**
Given:

products = ["Laptop", "Phone", "Tablet"]

prices = [50000, 20000, 15000]

quantities = [5, 10, 7]


## a. Compute total revenue for each product.

```python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

for i in range(len(products)):
  revenue=prices[i]*quantities[i]
  print(products[i],"Revenue:",revenue)
```

## b.Identify which product generated maximum revenue.

```python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

revenue=[prices[i]*quantities[i] for i in range(len(products))]
max_rev=max(revenue)
product=products[revenue.index(max_rev)]
print("Product with max revenue is", product)
print("Max revenue is", max_rev)
```

## c.Store revenues in a new list.

```python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

revenue=[]
for i in range(len(products)):
  revenue.append(prices[i]*quantities[i])

print("Store revenue:",revenue)
```

## d. Print a formatted summary like:

```python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

for i in range(len(products)):
  revenue=prices[i]*quantities[i]
  print(products[i],"Price:",prices[i],"Quantities:",quantities[i],"Revenues:",revenue)
```
